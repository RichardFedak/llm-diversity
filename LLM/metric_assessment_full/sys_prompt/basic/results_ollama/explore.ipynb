```python
import json
import os
```

```python
avg_counts = {
    "llama_basic" : {
        "A": 0,
        "B": 0,
        "C": 0,
        "other": 0
    },
    "llama_summary" : {
        "A": 0,
        "B": 0,
        "C": 0,
        "other": 0
    },
    "llama_standouts" : {
        "A": 0,
        "B": 0,
        "C": 0,
        "other": 0
    },
}

basic_files_count = 0
summary_files_count = 0
standouts_files_count = 0


for filename in os.listdir("."):
    if filename.endswith(".json") and not filename.startswith("_") and not filename.startswith("c"):
        with open(filename, "r") as file:
            data = json.load(file)
            if filename.startswith("summary"):
                for evaluation in data["evaluations"]:
                    if evaluation["output"] in avg_counts["llama_summary"]:
                        avg_counts["llama_summary"][evaluation["output"]] += 1
                    else:
                        avg_counts["llama_summary"]["other"] += 1
                    avg_counts["llama_summary"]
                summary_files_count += 1
            elif filename.startswith("standouts"):
                for evaluation in data["evaluations"]:
                    if evaluation["output"] in avg_counts["llama_standouts"]:
                        avg_counts["llama_standouts"][evaluation["output"]] += 1
                    else:
                        avg_counts["llama_standouts"]["other"] += 1
                standouts_files_count += 1
            else:
                for evaluation in data["evaluations"]:
                    if evaluation["output"] in avg_counts["llama_basic"]:
                        avg_counts["llama_basic"][evaluation["output"]] += 1
                    else:
                        avg_counts["llama_basic"]["other"] += 1
                basic_files_count += 1


# Calculate averages
for key in avg_counts:
    for output in avg_counts[key]:
        if key == "llama_basic":
            avg_counts[key][output] /= basic_files_count
        elif key == "llama_summary":
            avg_counts[key][output] /= summary_files_count
        elif key == "llama_standouts":
            avg_counts[key][output] /= standouts_files_count

print("Average counts for llama_basic:")
for output, count in avg_counts["llama_basic"].items():
    print(f"{output}: {count:.2f}")

print("\nAverage counts for llama_summary:")
for output, count in avg_counts["llama_summary"].items():
    print(f"{output}: {count:.2f}")

print("\nAverage counts for llama_standouts:")
for output, count in avg_counts["llama_standouts"].items():
    print(f"{output}: {count:.2f}")


```
