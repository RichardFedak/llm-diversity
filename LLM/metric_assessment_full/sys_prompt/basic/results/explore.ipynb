```python
import json
import os
```

```python
avg_counts = {
    "gemini_basic" : {
        "A": 0,
        "B": 0,
        "C": 0,
        "other": 0
    },
    "gemini_summary" : {
        "A": 0,
        "B": 0,
        "C": 0,
        "other": 0
    },
    "gemini_standouts" : {
        "A": 0,
        "B": 0,
        "C": 0,
        "other": 0
    },
}

basic_files_count = 0
summary_files_count = 0
standouts_files_count = 0


for filename in os.listdir("."):
    if filename.endswith(".json") and not filename.startswith("_") and not filename.startswith("c"):
        with open(filename, "r") as file:
            data = json.load(file)
            if filename.startswith("summary"):
                for evaluation in data["evaluations"]:
                    if evaluation["output"] in avg_counts["gemini_summary"]:
                        avg_counts["gemini_summary"][evaluation["output"]] += 1
                    else:
                        avg_counts["gemini_summary"]["other"] += 1
                    avg_counts["gemini_summary"]
                summary_files_count += 1
            elif filename.startswith("standouts"):
                for evaluation in data["evaluations"]:
                    if evaluation["output"] in avg_counts["gemini_standouts"]:
                        avg_counts["gemini_standouts"][evaluation["output"]] += 1
                    else:
                        avg_counts["gemini_standouts"]["other"] += 1
                standouts_files_count += 1
            else:
                for evaluation in data["evaluations"]:
                    if evaluation["output"] in avg_counts["gemini_basic"]:
                        avg_counts["gemini_basic"][evaluation["output"]] += 1
                    else:
                        avg_counts["gemini_basic"]["other"] += 1
                basic_files_count += 1


# Calculate averages
for key in avg_counts:
    for output in avg_counts[key]:
        if key == "gemini_basic":
            avg_counts[key][output] /= basic_files_count
        elif key == "gemini_summary":
            avg_counts[key][output] /= summary_files_count
        elif key == "gemini_standouts":
            avg_counts[key][output] /= standouts_files_count

print("Average counts for gemini_basic:")
for output, count in avg_counts["gemini_basic"].items():
    print(f"{output}: {count:.2f}")

print("\nAverage counts for gemini_summary:")
for output, count in avg_counts["gemini_summary"].items():
    print(f"{output}: {count:.2f}")

print("\nAverage counts for gemini_standouts:")
for output, count in avg_counts["gemini_standouts"].items():
    print(f"{output}: {count:.2f}")


```
